#### Univariate analysis

```python
# Distribution of Price
sns.distplot(df["Price"])
plt.title("Distribution of Price")
plt.show()
```

```python
plt.boxplot(df.Price)
plt.show()
# Many outliers, ie many observations are away from normal observation
# we can replace them with median or remove them while processing
```

```python
# Convert "number of floors" to integers, as some floors were marked in decimal points (eg 1.5) which can not be possible
df['number of floors'] = df['number of floors'].astype(int)

# Finding the number of unique floors
unique_floor_values = df['number of floors'].unique()
print(unique_floor_values)

# Distribution of the number of floors
floor_counts = df["number of floors"].value_counts()
floor_values = floor_counts.index
floor_occurrences = floor_counts.values

labels = [f"Floor {floor}: {count}" for floor, count in zip(floor_values, floor_occurrences)]

plt.pie(floor_occurrences,[0.02,0.05,0],labels=labels, autopct='%1.1f%%',colors=['red','blue','green'])
plt.title('Distribution of Number of Floors')
plt.show()

#Clearly there are maximum houses with 1 floors.
```

#### Bi - variate analysis

```python
sns.jointplot(x='number of bedrooms',y='Price',data=df)
```

```python
plt.scatter(df['living area'], df['Price'])
plt.xlabel('Living Area')
plt.ylabel('Price')
plt.title('Living Area vs Price')
plt.show()
```

```python
sns.scatterplot(x='number of bathrooms',y='Price',data=df)
plt.xlabel('Number of bathrooms')
plt.ylabel('Price')
plt.title('Number of bathrooms vs Price')
plt.show()
```

```python
mean_price_by_floors = df.groupby('number of floors')['Price'].mean()
mean_price_by_floors.plot(kind='bar')
plt.xlabel('Number of floors')
plt.ylabel('Mean price')
plt.title('Mean price by number of floors')
plt.show()
#mean price of 2 floors is maximum
```

```python
sns.barplot(x='number of bedrooms',y='Price',data=df)
plt.xlabel('Number of bedrooms')
plt.ylabel('Mean price')
plt.title('Number of bedrooms by Price')
plt.show()
```
